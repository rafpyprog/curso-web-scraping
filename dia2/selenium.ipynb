# Selenium
Selenium utiliza um driver para conectar-se a um browser e controlá-lo de maneira automatizada.

## Instalação do pacote
``` pip install selenium ```

## Onde fazer o download dos drivers
Cada browser possui um driver específico. O arquivo deve estar no PATH da máquina ou na mesma pasta do script Python.
* Chrome: 	https://sites.google.com/a/chromium.org/chromedriver/downloads
* Edge: 	https://developer.microsoft.com/en-us/microsoft-edge/tools/webdriver/
* Firefox: 	https://github.com/mozilla/geckodriver/releases
* Safari: 	https://webkit.org/blog/6900/webdriver-support-in-safari-10/

## Quando usar Selenium
Quando não for possível simular a iteração do usuário através do Requests (Ex: reCaptcha) ou em sites cujo conteúdo é  carregado através de Javascript, necessitande de um browser para renderização.

## Iniciando o Browser

````python
import time
from IPython.display import Image
from selenium.webdriver import Chrome, Firefox

driver = Chrome()  #Firefox()
url = 'https://br.financas.yahoo.com/'
driver.get(url)
````

## Navegando

````python
driver.get('http://www.google.com')
time.sleep(2)
driver.back()
````

## Localizando um elemento

````python
campo_busca = driver.find_element_by_name('p')
````

## Enviando comandos e pressionando teclas

````python
from selenium.webdriver.common.keys import Keys

campo_busca.send_keys('PETR3')
campo_busca.send_keys(Keys.ENTER)
````

## Screenshot

````python
img = driver.get_screenshot_as_png()
Image(img)
````

````python
estatisticas = driver.find_element_by_xpath('//*[@id="quote-nav"]/ul/li[4]/a/span')
estatisticas.click()
````

````python
xpath_tabela_medidas = '//*[@id="Col1-0-KeyStatistics-Proxy"]/section/div[2]/div[1]/div[1]/div/table'
tabela = driver.find_element_by_xpath(tabela_medidas)
````

````python
import pandas as pd

pd.read_html(tabela.get_attribute('outerHTML'))[0]
````
